```python
df = raw_df.copy()
```

```python
# Используем стандартизацию
standartize_columns = [
    'Latitude',
    'Longitude',
    'Depth',
    'Thickness (gross average ft)',
    'Thickness (net pay average ft)',
    'Porosity',
    'Permeability'
]

for label in standartize_columns:
  scaler = StandardScaler()
  df[label] = scaler.fit_transform(df[[label]])
```

```python
# Преобразуем категориальные переменные в численные

to_int = [
    'Field name',
    'Reservoir unit',
    'Region',
    'Hydrocarbon type',
    'Reservoir status',
    'Reservoir period',
    'Lithology'
]

to_int_multy = ['Country', 'Tectonic regime', 'Operator company', 'Basin name', 'Structural setting']

for label in to_int:
  enc = OneHotEncoder(sparse_output=False, handle_unknown='ignore')
  encoded = enc.fit_transform(df[[label]])
  encoded_df = pd.DataFrame(encoded, columns=enc.get_feature_names_out([label]))
  df = pd.concat([df, encoded_df], axis=1)
  df = df.drop(label, axis=1)

y_enc = LabelEncoder()
df['Onshore/Offshore'] = y_enc.fit_transform(df['Onshore/Offshore'])
```

```python
for label in to_int_multy:
  print(label, raw_df[label].unique())
  label_s = f'{label}_split'
  df[label_s] = df[label].str.split('/')

  enc = MultiLabelBinarizer()
  print(df[label_s])

  encoded = enc.fit_transform(df[label_s])
  encoded_df = pd.DataFrame(encoded, columns=enc.classes_)
  df = pd.concat([df, encoded_df], axis=1)
  df = df.drop(label, axis=1)
  df = df.drop(label_s, axis=1)
```

```python
df.head()
```

```python
df.dropna(inplace=True)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(['Onshore/Offshore'], axis=1)
y = df['Onshore/Offshore']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
model = LogisticRegression().fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()
print("TN:", tn)
print("FP:", fp)
print("FN:", fn)
print("TP:", tp)
print("\nAccuracy:", accuracy_score(y_test, y_pred))
print("Recall:", recall_score(y_test, y_pred))
print("Precision:", precision_score(y_test, y_pred))
print("F1:", f1_score(y_test, y_pred))
```
