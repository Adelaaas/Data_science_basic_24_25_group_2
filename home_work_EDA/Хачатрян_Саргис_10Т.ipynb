## График 6

Постройте график среднегодовой температуры, т.е. график зависимости года от средней температуры в этом году.

```python
avg_t = df.groupby('Year')['Temperature (C)'].mean().reset_index()

plt.title('Average Temperature Graph')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.xticks(ticks=range(avg_t['Year'].min(), avg_t['Year'].max() + 1, 1), rotation=45)
sns.lineplot(data=avg_t, x='Year', y='Temperature (C)')
```

Выводы:
1. 2005 был аномально холодным годом
2. с 2006 года температура держалась на приблизительно одинаковом уровне ~11.75C

## График 7

Найдите любой **нестандартный** график и постройте его. На ваш выбор.

Вариант расширения функции Вейерштрасса в 3D:
$$\large z=\sum_{n=0}^{\infty} b^{n} \cos(a^{n} \pi xy)$$
("Математический монстр" - непрерывная функция нигде не имеющая производную)

```python
N = 20
a = 3
b = 0.5

xx = np.linspace(-2, 2, 250)
yy = np.linspace(-2, 2, 250)
xx, yy = np.meshgrid(xx, yy)

z = np.sum([b**n * np.cos(a**n * np.pi * xx * yy) for n in range(N)], axis=0)

fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(xx, yy, z, cmap='viridis', edgecolor='none')

ax.set_title(r"$z = \sum_{n=0}^{\infty} b^n \cos(a^n \pi xy)$", fontsize=16)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

fig.colorbar(surf)
plt.tight_layout()
plt.show()
```
