```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Надо будет изучить информациюю хранящююся в датасете sales_dataset.csv. Он содержит информацию о ценах на жилье с характеристиками этого жилья.

Дальше надо указать путь к нашему датасету

```python
data = pd.read_csv("sales_dataset.csv")
data
```

1ый График

```python
colors = {'For Sale': 'blue', 'For Rent': 'orange'}

plt.figure(figsize=(12, 8))
sns.scatterplot(data=data, x='Area_in_Marla', y='price', hue='purpose',
                palette=colors, alpha=0.5)

plt.title('Зависимость цены от площади (Area_in_Marla) с учетом цели')
plt.xlabel('Площадь (в марлах)')
plt.ylabel('Цена (в рупиях)')
plt.legend(title='Цель')
plt.grid(True)
plt.show()
```

2ой График

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 10))
fig.patch.set_facecolor('#f0f0f0')


axs[0, 0].hist(data['property_type'], color='lightblue', edgecolor='black')
axs[0, 0].set_title('Распределение по типу недвижимости')
axs[0, 0].set_xlabel('Тип недвижимости')
axs[0, 0].set_ylabel('Количество объектов')

axs[0, 1].hist(data['city'], color='lightgreen', edgecolor='black')
axs[0, 1].set_title('Распределение по городам')
axs[0, 1].set_xlabel('Город')
axs[0, 1].set_ylabel('Количество объектов')


axs[1, 0].hist(data['purpose'], color='salmon', edgecolor='black')
axs[1, 0].set_title('Распределение по цели')
axs[1, 0].set_xlabel('Цель')
axs[1, 0].set_ylabel('Количество объектов')


axs[1, 1].hist(data['bedrooms'], color='lightcoral', edgecolor='black')
axs[1, 1].set_title('Распределение по количеству спален')
axs[1, 1].set_xlabel('Количество спален')
axs[1, 1].set_ylabel('Количество объектов')


plt.tight_layout()
plt.show()
```

3ий График

```python
numeric_data = data[['price', 'baths', 'bedrooms', 'Area_in_Marla']]

correlation_matrix = numeric_data.corr()


plt.figure(figsize=(10, 8))
sns.set(font_scale=1.2)
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm',
                       linewidths=.5)


plt.title('Матрица корреляции', fontsize=16)
plt.xticks(rotation=45)
plt.yticks(rotation=45)
plt.tight_layout()
plt.show()
```

4ый График

```python
plt.figure(figsize=(14, 12))

colors = {'For Sale': 'blue', 'For Rent': 'orange'}

plt.subplot(2, 1, 1)
sns.boxplot(x='purpose', y='price', data=data, palette=colors)
plt.title('Boxplot цены по цели')
plt.xlabel('Цель')
plt.ylabel('Цена (в рупиях)')


plt.subplot(2, 1, 2)
sns.boxplot(x='purpose', y='bedrooms', data=data, palette=colors)
plt.title('Boxplot количества спален по цели')
plt.xlabel('Цель')
plt.ylabel('Количество спален')


plt.tight_layout()
plt.show()
```

5ый График

```python
missing_data = data.isnull()


plt.figure(figsize=(12, 8))
sns.heatmap(missing_data, cmap='coolwarm', cbar=False, yticklabels=False)


plt.title('Heatmap пропусков в данных', fontsize=16)
plt.xlabel('Переменные')
plt.ylabel('Записи')
plt.tight_layout()
plt.show()
```

Выводы: На графике не видно наличие пропусков в данных. Это важно учитывать при дальнейшей обработке.

6ой График

```python
average_price_by_season = data.groupby('Season')['price'].mean().reset_index()

plt.figure(figsize=(10, 6))
plt.bar(average_price_by_season['Season'], average_price_by_season['price'], color='skyblue')
plt.title('Среднее значение цены в зависимости от сезона')
plt.xlabel('Сезон')
plt.ylabel('Средняя цена (в рупиях)')
plt.xticks(average_price_by_season['Season'])
plt.grid(True)
plt.tight_layout()
plt.show()
```
