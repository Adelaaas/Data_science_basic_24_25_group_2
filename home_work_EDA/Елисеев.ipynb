Необходимо изучить информацию хранящаюся в датасете weatherHistory.csv. Он содержит информацию о погоде.

```python
df = pd.read_csv("/content/weatherHistory_var1.csv")
df
```

График №1

Построить точечный график зависимости переменной Temperature (C) от Wind Speed (km/h) с учётом третей переменной Precip Type.

```python
plt.title('Dependence of Temperature(C) on Wind Speed(km/h)')
plt.xlabel('Wind Speed(km/h)')
plt.ylabel('Dependence of Temperature(C)')
sb.scatterplot(data=df, x='Wind Speed (km/h)', y='Temperature (C)', hue='Precip Type', palette="colorblind", s=30)
plt.show()
```

Вывод: в снежную погоду температура меньше чем в дождливую

График №2



Постройте 4 графика на одной фигуре, как расположить координатные плоскости - это на ваше усмотрение.

Это должны быть гистограммы для следующих переменных:

    Precip Type
    Season
    DayOfWeek
    Year

Каждый график должен быть своего цвета, также задайте цвет отличный от белого для бекграунда всей фигуры.


```python
plt.subplots_adjust(wspace=0.6, hspace=0.6)

plt.subplot(221)
plt.title('"Precip Type" histogram')
sb.histplot(df['Precip Type'])

plt.subplot(222)
plt.title('"Season"(Quarter) histogram')
plt.xticks(np.arange(1, 5, 1))
sb.histplot(df['Season'], discrete=True)

plt.subplot(223)
plt.title('"Day of Week" histogram')
plt.xlabel('Count')
sb.barplot(df['DayOfWeek'])

plt.subplot(224)
plt.title('"Year" histogram')
sb.histplot(df['Year'], discrete=True)

plt.show()
```

Вывод: Снежных дней меньше, чем дождливых. Данных для каждого дня недели, квартала, года находятся в приблизительно одинаковом количестве

График №3



Визуализируйте матрицу корреляции в виде heatmap.

Все должно быть ЧИТАБЕЛЬНО!


```python
plt.figure(figsize=(12,10))

corr_matrix = df.corr(numeric_only=True)
sb.heatmap(corr_matrix, xticklabels=df.corr(numeric_only=True).columns, yticklabels=df.corr(numeric_only=True).columns, cmap='Blues', center=0, annot=True)
plt.title('Correlogram of mtcars', fontsize=22)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

Вывод: влажность и температура имеют сильную отрицательную корелляцию. Видимость имеет некоторую отрицательную корелляцию с влажностью и некоторую положительную корелляцию с температурой.

График №5

Постройте график распределения переменной Wind Bearing (degrees) с учетом переменной Precip Type.

```python
plt.title('Wind Bearing (degrees) Distribution Graph')
sb.kdeplot(data=df, x='Wind Bearing (degrees)', hue='Precip Type')
```

Вывод: 1.Волнообразное распределение: Общее распределение характеризуется волнообразной структурой, что указывает на нестабильность направления. Это может предполагать периодическую природу данных. 2.В дождливую погоду направление ветра менее стабильно чем в снежную.

График №6

Постройте график среднегодовой температуры, т.е. график зависимости года от средней температуры в этом году.

```python
avg_t = df.groupby('Year')['Temperature (C)'].mean().reset_index()

plt.title('Average Temperature Graph')
plt.xlabel('Year')
plt.ylabel('Average Temperature')
plt.xticks(ticks=range(avg_t['Year'].min(), avg_t['Year'].max() + 1, 1), rotation=45)
sb.lineplot(data=avg_t, x='Year', y='Temperature (C)')
```

Вывод: 1.2005 был аномально холодным годом. 2.С 2006 года температура держалась на приблизительно одинаковом уровне ~11.75C

График №7

Найдите любой нестандартный график и постройте его. На ваш выбор

```python
labels = ['Категория A', 'Категория B', 'Категория C', 'Категория D']
sizes = [15, 30, 45, 10]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen']
explode = (0.1, 0, 0, 0)

plt.figure(figsize=(8, 8))
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=140)

plt.axis('equal')
plt.title('Круговая диаграмма распределения категорий', fontsize=16)
plt.show()
```
