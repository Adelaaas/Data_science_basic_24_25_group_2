```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Надо будет изучить информациюю хранящююся в датасете pollution_dataset.csv.
Он содержит информацию о качестве воздуха и связанных с ним параметрах.

Дальше надо указать путь к нашему датасету


```python
data = pd.read_csv("/content/pollution_dataset.csv")
data
```

График 1

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x='CO', y='NO2', hue='Air Quality')
plt.title('Зависимость NO2 от CO с учетом качества воздуха')
plt.xlabel('Концентрация CO')
plt.ylabel('Концентрация NO2')
plt.legend(title='Качество воздуха')
plt.grid(True)
plt.show()
```

Выводы:
На графике видно, что с увеличением концентрации CO наблюдается рост уровня NO2.
Однако, качество воздуха также влияет на распределение точек, что указывает на возможные зависимости.также мы видим чем хуже вохдвух тем больше в нем NO2 Co

График 2


```python
fig, ax = plt.subplots(2, 2, figsize=(15, 15))


ax[0, 0].hist(data["Air Quality"], color='orange')
ax[0, 0].set_title("Гистограмма уровня качества воздуха")
ax[0, 0].set_xlabel("Качество воздуха")
ax[0, 0].set_ylabel("Частота")

sns.distplot(data["Population_Density"], color='green', ax=ax[0, 1])
ax[0, 1].set_title("График распределения плотности населения")
ax[0, 1].set_xlabel("Плотность населения (чел./км²)")

sns.distplot(data["Temperature"], color='red', ax=ax[1, 0])
ax[1, 0].set_title("График распределения температуры")
ax[1, 0].set_xlabel("Температура (°C)")

sns.distplot(data["Humidity"], ax=ax[1, 1])
ax[1, 1].set_title("График распределения влажности")
ax[1, 1].set_xlabel("Влажность (%)")

plt.show()
```

Выводы:
Гистограмма показывает преобладание умеренного качества воздуха.
Плотность населения и температура имеют широкий разброс значений.Влажность варьируется в пределах от низких до высоких значений.

График 3

```python
plt.figure(figsize=(12, 12))
correlation_matrix = data.corr(numeric_only=True)
sns.heatmap(correlation_matrix, xticklabels=correlation_matrix.columns, yticklabels=correlation_matrix.columns, cmap='Reds', center=0, annot=True)
plt.title("Матрица корреляции")
plt.show()
```

Выводы:
Корреляционная матрица показывает сильную положительную корреляцию между PM2.5 и PM10.
Также имеется отрицательная корреляция между CO и NO2.

График 4

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(x='Air Quality', y='CO', data=data)
plt.title('Boxplot CO по Air Quality')

plt.subplot(1, 2, 2)
sns.boxplot(x='Air Quality', y='NO2', data=data)
plt.title('Boxplot NO2 по Air Quality')


plt.show()
```


Выводы:
Boxplot показывает различия в уровнях CO и NO2 в зависимости от качества воздуха.
Более высокие уровни загрязнения соответствуют плохому качеству воздуха.

График 5


```python
plt.figure(figsize=(10, 6))
sns.heatmap(data.isnull(), cbar=False)
plt.title('Отображение пропусков в данных')
plt.show()
```

Выводы:
На графике не видно наличие пропусков в данных. Это важно учитывать при дальнейшей обработке.

График 6

```python
mean_pm10 = data.groupby('Air Quality')['PM10'].mean().reset_index()
plt.figure(figsize=(8,6))
sns.barplot(x='Air Quality', y='PM10', data=mean_pm10)
plt.title('Среднее значение PM10 в зависимости от качества воздуха')
plt.xlabel('Качество воздуха')
plt.ylabel('Средняя концентрация PM10 (мкг/м³)')
plt.show()
```

Выводы:
Средние значения PM10 показывают рост концентрации с ухудшением качества воздуха.

График 7

```python
air_quality_counts = data['Air Quality'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(air_quality_counts, labels=air_quality_counts.index, autopct='%1.1f%%')
plt.title('Распределение Air Quality')
plt.axis('equal')
plt.show()
```

В круговой диаграмме распределения качества воздуха видно, что большинство значений относится к категориям "Хорошее" и "Умеренное", в то время как "Опасное" качество воздуха встречается реже, что указывает на преобладание приемлемых условий в большинстве регионов.
